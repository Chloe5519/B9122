```python
def EMI(rate,N,PV,FV):
    result=(PV+FV/(1+rate)**N)*rate*(1+rate)**N/((1+rate)**N-1)
    return result
```

```python
readingFromFile = open(path,'r')
```

```python
contentFromFile=readingFromFile.read().lower()
readingFromFile.close()
contentList=contentFromFile.split("\n")
```

```python
print(contentList)
```

```python
words=[]
for sentence in contentList:
    words.extend(sentence.split(' '))
```

```python
wordsCount={}
for item in words:
    if item in wordsCount:
        wordsCount[item]+=1
    else:
        wordsCount[item]=1
```

```python
wordsCount_sorted = sorted(wordsCount.items(), reverse = True, key = lambda x:x[1])
```

```python
output=open("result.txt","w",encoding="utf-8")
for answer in wordsCount_sorted:
    print(answer[0],answer[1],sep=' ',file=output)
```

## Question 3

```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
seed_url="http://www8.gsb.columbia.edu/"
```

```python
urls = {seed_url:1}   #queue of urls to crawl
seen = {} # stack of urls seen so far
opened = []
keywords=['finance', 'engineering', 'business', 'research']
maxNumUrl = 50 #set the maximum number of urls to visit
```

```python
while len(urls) > 0 and len(opened) < maxNumUrl:
    # DEQUEUE A URL FROM urls AND TRY TO OPEN AND READ IT
    try:
        curr_url=max(urls,key=urls.get)
        del urls[curr_url] 
        print("num. of URLs in stack: %d " % len(urls))
        webpage=urllib.request.urlopen(curr_url)
        opened.append(curr_url)

    except Exception as ex:
        print(ex)
        continue    #skip code below

    # IF URL OPENS, CHECK WHICH URLS THE PAGE CONTAINS
    # Add this url with its score to seen
    soup = BeautifulSoup(webpage)  #creates object soup
    htmltext=soup.get_text().lower()
    
    stat=[]
    for item in keywords:
        num=htmltext.count(item)
        stat.append(num)
    score=sum(stat)
    seen[curr_url]=score
    
    
    if score>0:
    # Put child URLs into the stack
        for tag in soup.find_all('a', href = True): #find tags with links
            childUrl = tag['href']          #extract just the link
            o_childurl = childUrl
            childUrl = urllib.parse.urljoin(seed_url, childUrl)
            print("url=" + seed_url)
            print("original childurl=" + o_childurl)
            print("childurl=" + childUrl)
            print("url in childUrl=" + str(seed_url in childUrl))
            print("childUrl not in seen=" + str(childUrl not in seen))
            if seed_url in childUrl and childUrl not in seen:
                print("***urls.append and seen.append***")
                urls[childUrl]=score
            else:
                print("######")
```

```python
print("The number of URLs seen: "+str(len(seen))+" and opened: "+str(len(opened)))
```

```python
seen_sorted = sorted(seen.items(), reverse = True, key = lambda x:x[1])
print("Top 25 scoring URLs seen:")
print(seen_sorted[:25])
```
